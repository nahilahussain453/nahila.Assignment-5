```python
#Linear relation between year and price
x=data['Year']
y=data['Price']
print(data['Year'].values)
```

```python
x=(data['Year'].values.reshape(-1,1))
y=(data['Price'].values.reshape(-1,1))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print("x_train is\n",x_train)
print("y_train is\n",y_train)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#prediction
pred=regressor.predict([[10]])
print("for given year price is",pred)
```

```python
y_pred=regressor.predict(x_test)
print("y_pred for given data is",y_pred)
```

```python
import numpy
x_test=list(numpy.concatenate(x_test).flat)
x_test
y_test=list(numpy.concatenate(y_test).flat)
y_test
y_pred=list(numpy.concatenate(y_pred).flat)
y_pred
```

```python
data_new=pd.DataFrame({"Years":x_test,"Actual data":y_test,"Predicted":y_pred})
data_new
```

```python
#MODEL EVALUATION
#MEAN SQUARED ERROR
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(y_test,y_pred)
print("The MSe for this model prediction is",mse)
print("The r squared value is ",r2_score(y_test,y_pred))
```
